```python
def limit_quantiles(df, column, lower_quantile=0.1, upper_quantile=0.9):
    df_limited = df.copy()
    quantiles = df[column].quantile([lower_quantile, upper_quantile])
    lower_bound = quantiles.loc[lower_quantile]
    upper_bound = quantiles.loc[upper_quantile]
    df_limited[column] = df_limited[column].clip(lower=lower_bound, upper=upper_bound)
    return df_limited



```

### Summary of Training of Different Unsupervised Models

In this analysis, various unsupervised learning models were applied to the stock market dataset to uncover hidden patterns and structures. The models used include:

#### 1. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
- **Objective**: To identify clusters of varying shapes and sizes, and to detect outliers.
- **Parameters**: `eps=0.5`, `min_samples=15`
- **Results**: The DBSCAN algorithm identified clusters and labeled outliers. The silhouette score was calculated to evaluate the quality of the clusters.

#### 2. KMeans Clustering
- **Objective**: To partition the dataset into a predefined number of clusters.
- **Parameters**: `n_clusters=4` (determined using the elbow method)
- **Results**: The KMeans algorithm assigned each data point to one of the 4 clusters. The silhouette score was calculated to evaluate the quality of the clusters.

#### 3. Hierarchical Clustering
- **Objective**: To create a hierarchy of clusters using a dendrogram.
- **Method**: Ward's linkage method
- **Results**: The hierarchical clustering algorithm produced a dendrogram, and clusters were formed based on a distance threshold. The silhouette score was calculated to evaluate the quality of the clusters.

### Evaluation Metrics
- **Silhouette Score**: Used to measure how similar an object is to its own cluster compared to other clusters. Higher scores indicate better-defined clusters.
- **Visualization**: Scatter plots of the UMAP projections were used to visualize the clusters formed by each algorithm.

### Conclusion
Each clustering algorithm provided different insights into the structure of the dataset. DBSCAN was effective in identifying outliers, KMeans provided a clear partitioning of the data, and hierarchical clustering offered a hierarchical view of the clusters. The silhouette scores helped in assessing the quality of the clusters formed by each algorithm.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Aplicar StandardScaler
scaler_standard = StandardScaler()
scale_data = pd.DataFrame(scaler_standard.fit_transform(numeric_df))

# Aplicar MinMaxScaler
scaler_minmax = MinMaxScaler()
scale_data = pd.DataFrame(scaler_minmax.fit_transform(scale_data))
```

```python
import umap

# Reducir la dimensionalidad a 2 dimensiones usando UMAP
reducer = umap.UMAP(n_components=2,random_state=42,init='pca')
embedding = reducer.fit_transform(scale_data)

# Crear un DataFrame con las nuevas dimensiones
embedding_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])

# Visualizar los resultados
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', data=embedding_df)
plt.title('UMAP projection of the scaled data')
plt.show()
```

```python
# Calcular la relevancia de cada variable en función de la correlación con las dimensiones UMAP
correlations = {}
for col in numeric_df.columns:
    corr_umap1 = np.corrcoef(numeric_df[col], embedding_df['UMAP1'])[0, 1]
    corr_umap2 = np.corrcoef(numeric_df[col], embedding_df['UMAP2'])[0, 1]
    relevance = abs(corr_umap1) + abs(corr_umap2)
    correlations[col] = relevance

# Ordenar las variables por relevancia
sorted_features = sorted(correlations, key=correlations.get, reverse=True)

# Seleccionar las 3 variables más relevantes
top_3_features = sorted_features[:3]

# Añadir las 3 variables más relevantes al DataFrame de embedding
for f in top_3_features:
    embedding_df[f] = numeric_df[f].values

# Visualizar la proyección UMAP coloreada por las 3 variables más relevantes
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, f in enumerate(top_3_features):
    sns.scatterplot(
        x='UMAP1',
        y='UMAP2',
        hue=f,
        data=embedding_df,
        palette='viridis',
        ax=axes[i]
    )
    axes[i].set_title(f'UMAP coloreado por {f}')
    axes[i].legend(title=f, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=15)
clusters = dbscan.fit_predict(embedding)

# Añadir los clusters al DataFrame de embedding
embedding_df['Cluster'] = clusters

# Visualizar los resultados de la clusterización
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='Cluster', palette='viridis', data=embedding_df)
plt.title('DBSCAN Clustering on UMAP projection')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm
import numpy as np
import matplotlib.pyplot as plt

# Filtrar puntos con clusters válidos (diferentes a -1) para la silueta
# Si deseas incluir el cluster -1 (ruido), debes tener en cuenta que la silueta
# no está definida para un solo cluster o para outliers sin asignación.
valid_mask = (clusters != -1)
X_valid = embedding[valid_mask]
labels_valid = clusters[valid_mask]

# Calcular el coeficiente de silueta global
silhouette_avg = silhouette_score(X_valid, labels_valid)
print("Silhouette Score:", silhouette_avg)

# Calcular las muestras de silueta
sample_silhouette_values = silhouette_samples(X_valid, labels_valid)

# Crear la figura para la gráfica de silueta
fig, ax1 = plt.subplots(figsize=(8, 6))

# El eje y (vertical) quedará vacío, sólo necesitaremos el x (horizontal)
y_lower = 10
unique_clusters = np.unique(labels_valid)

# Para asignar un color a cada cluster
colors = cm.nipy_spectral(np.linspace(0, 1, len(unique_clusters)))

for i, c in enumerate(unique_clusters):
    # Obtener las muestras de silueta que pertenecen al cluster c
    ith_cluster_silhouette_values = sample_silhouette_values[labels_valid == c]

    # Ordenamos las muestras de silueta por su valor
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    # Rellenar la gráfica con los valores de silueta del cluster
    ax1.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values,
        facecolor=colors[i],
        edgecolor=colors[i],
        alpha=0.7
    )

    # Etiquetar el cluster en el gráfico
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(c))

    # Actualizar la posición inferior para el siguiente cluster
    y_lower = y_upper + 10

# Configurar el eje X con el rango de la silueta (-1 a 1)
ax1.set_xlim([-0.1, 1])
# El eje Y será dinámico según el total de puntos
ax1.set_ylim([0, y_lower + 10])
ax1.set_xlabel("Valor de silueta")
ax1.set_ylabel("Cluster")

# Línea vertical mostrando el promedio de silueta
ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

plt.title("Gráfico de la Silueta para DBSCAN")
plt.show()
```

```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

# Calcular la suma de los errores cuadráticos (SSE) para diferentes valores de k
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(embedding)
    sse.append(kmeans.inertia_)

# Graficar el codo
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('SSE')
plt.title('Método del codo para determinar el número óptimo de clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Aplicar KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_clusters = kmeans.fit_predict(embedding)

# Añadir los clusters al DataFrame de embedding
embedding_df['KMeans_Cluster'] = kmeans_clusters

# Visualizar los resultados de la clusterización
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='KMeans_Cluster', palette='viridis', data=embedding_df)
plt.title('KMeans Clustering on UMAP projection')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples

import matplotlib.cm as cm

def plot_silhouette(data, n_clusters):
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(18, 7)

    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(data) + (n_clusters + 1) * 10])

    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data)

    silhouette_avg = silhouette_score(data, cluster_labels)
    print(f'For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}')

    sample_silhouette_values = silhouette_samples(data, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    ax1.set_title(f'The silhouette plot for the various clusters (n_clusters = {n_clusters})')
    ax1.set_xlabel('The silhouette coefficient values')
    ax1.set_ylabel('Cluster label')

    ax1.axvline(x=silhouette_avg, color='red', linestyle='--')
    ax1.set_yticks([])
    ax1.set_xticks(np.arange(-0.1, 1.1, 0.2))

    plt.show()

# Plot silhouette for 4 clusters
plot_silhouette(embedding, 4)

# Plot silhouette for 5 clusters
plot_silhouette(embedding, 5)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Calcular el enlace jerárquico
Z = linkage(embedding, method='ward')

# Graficar el dendrograma
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Dendrograma de Clustering Jerárquico')
plt.xlabel('Índice de Muestra')
plt.ylabel('Distancia')
plt.show()

# Obtener los clusters
max_d = 10  # Puedes ajustar este valor según sea necesario
clusters_hierarchical = fcluster(Z, max_d, criterion='distance')

# Añadir los clusters al DataFrame de embedding
embedding_df['Hierarchical_Cluster'] = clusters_hierarchical

# Visualizar los resultados de la clusterización jerárquica
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='Hierarchical_Cluster', palette='viridis', data=embedding_df)
plt.title('Hierarchical Clustering on UMAP projection')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.cm as cm

def plot_silhouette_hierarchical(data, cluster_labels):
    # Filtrar los datos y etiquetas de los clusters que quedaron
    mask = cluster_labels != -1
    data_filtered = data[mask]
    cluster_labels_filtered = cluster_labels[mask]
    
    n_clusters = len(set(cluster_labels_filtered))
    fig, ax1 = plt.subplots(1, 1)
    fig.set_size_inches(18, 7)

    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(data_filtered) + (n_clusters + 1) * 10])

    silhouette_avg = silhouette_score(data_filtered, cluster_labels_filtered)
    print(f'For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}')

    sample_silhouette_values = silhouette_samples(data_filtered, cluster_labels_filtered)

    y_lower = 10
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels_filtered == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    ax1.set_title(f'The silhouette plot for the various clusters (n_clusters = {n_clusters})')
    ax1.set_xlabel('The silhouette coefficient values')
    ax1.set_ylabel('Cluster label')

    ax1.axvline(x=silhouette_avg, color='red', linestyle='--')
    ax1.set_yticks([])
    ax1.set_xticks(np.arange(-0.1, 1.1, 0.2))

    plt.show()

# Plot silhouette for hierarchical clustering
plot_silhouette_hierarchical(embedding, clusters_hierarchical)
```
